```python
import requests
import pandas as pd
from bs4 import BeautifulSoup

Product_name = []
Price = []
Description = []
Reviews = []

for i in range(2, 22):
    url1 = 'https://www.flipkart.com'
    url = 'https://www.flipkart.com/search?q=mobiles&as=on&as-show=on&otracker=AS_Query_TrendingAutoSuggest_1_0_na_na_na&otracker1=AS_Query_TrendingAutoSuggest_1_0_na_na_na&as-pos=1&as-type=HISTORY&suggestionId=mobiles&requestId=e24e027b-9f2b-4964-bfa3-e56edb4f2a56&as-backfill=on&page=' + str(i)
    r = requests.get(url)
    soup = BeautifulSoup(r.text, "lxml")
    box = soup.find("div", class_ = "_1YokD2 _3Mn1Gg")

    nxt = soup.find("a", class_="_1LKTO3").get("href")
    cnxt = url1 + nxt
    print(cnxt)

    names = box.find_all("div", class_="_4rR01T")
    for n in names:
        Product_name.append(n.text)

    prices = box.find_all("div", class_="_30jeq3 _1_WHN1")
    for p in prices:
        Price.append(p.text)

    desp = box.find_all("ul", class_="_1xgFaf")
    for d in desp:
        Description.append(d.text)

    rev = box.find_all("div", class_="_3LWZlK")
    for r in rev:
        Reviews.append(r.text)

df = pd.DataFrame({"Product_name": Product_name, "Price": Price, "Description": Description, "Reviews": Reviews})
print(df)
df.to_csv("MOBILEM.csv")
```
