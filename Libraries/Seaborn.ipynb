## What are the black lines on the bar ?

the black lines on the top middle of the bars typically represent error bars. Error bars provide a visual representation of the variability or uncertainty in the data being represented by each bar.

Error bars are commonly used to show the range of uncertainty around the mean or average value of each bar. They can convey information about the variability of the data points or the level of confidence in the measurement.

The length of the error bars can vary based on the type of error represented. Commonly used error bars include:

Standard Error (SE): The error bars represent the standard error of the mean, which provides a measure of how much the sample means are expected to vary from the true population mean.

Confidence Intervals (CI): The error bars represent the confidence intervals, typically at 95% confidence level, which indicates the range within which the true population mean is likely to lie.

Standard Deviation (SD): The error bars represent the standard deviation, which measures the spread of the data points around the mean.

Custom Error Values: You can also calculate and plot custom error values based on your specific analysis or experiment.

Error bars are useful for comparing different bar heights and understanding the uncertainty associated with each bar's value. They provide a more complete picture of the data distribution and can help in drawing meaningful conclusions from the graph.

Keep in mind that error bars are not always present in bar graphs. Whether to include error bars depends on the specific context of the data and the purpose of the visualization. In many cases, you can control the display of error bars using plotting functions in libraries like Matplotlib or Seaborn by specifying the appropriate parameters.


```python
sns.set(style = "darkgrid")
sns.barplot(x='island',y='bill_length_mm',hue='sex',data=new,hue_order=["Female","Male"],ci=100,n_boot=5,saturation=150,errcolor="r",capsize=0.37)
plt.show()
```

## ViolinPlot

A violin plot plays a similar role as box and whiskers ploy . It shows the distribution of a quantitative data across several levels of one (or more) categorical variables such that those distribution can be conpared

Unlike a box plot, in which all of the plot components correspond to actual datapoints, the violin plot features a kernel density estimation of the underlying distribution.

This can be an effective and attractive way to show multiple distributions of data at once, but keep in mind that the estimation procedure is influenced by the sample size, and violins for relatively small samples might look misleadingly smooth.

**SYNTAX :-**

seaborn.violinplot(data=None, *, x=None, y=None, hue=None, order=None, hue_order=None, bw='scott', cut=2, scale='area', scale_hue=True, gridsize=100, width=0.8, inner='box', split=False, dodge=True, orient=None, linewidth=None, color=None, palette=None, saturation=0.75, ax=None, **kwargs)

## PairPlot

Plot pairwise relationships in a dataset.

By default, this function will create a grid of Axes such that each numeric variable in data will by shared across the y-axes across a single row and the x-axes across a single column. The diagonal plots are treated differently: a univariate distribution plot is drawn to show the marginal distribution of the data in each column.

It is also possible to show a subset of variables or plot different variables on the rows and columns.

This is a high-level interface for PairGrid that is intended to make it easy to draw a few common styles. You should use PairGrid directly if you need more flexibility.

**SYNTAX :-**

seaborn.pairplot(data, *, hue=None, hue_order=None, palette=None, vars=None, x_vars=None, y_vars=None, kind='scatter', diag_kind='auto', markers=None, height=2.5, aspect=1, corner=False, dropna=False, plot_kws=None, diag_kws=None, grid_kws=None, size=None)

```python
sns.pairplot(tip,vars=['total_bill','tip'],hue="sex",palette='hsv')
plt.show()
```

```python
sns.pairplot(tip,hue="sex",palette='hsv',y_vars=["total_bill","tip"])
```

```python
sns.pairplot(tip,hue="sex",palette='hsv',kind="kde") #scatter,kde,hist,reg within kind
```

```python
sns.pairplot(tip,hue="sex",palette='hsv',kind="kde",diag_kind="hist") #scatter,kde,hist,reg within kind
```

## CatPlot

**Categorical scatterplots:**

swarmplot() (with kind="swarm")

**Categorical distribution plots:**

boxplot() (with kind="box")

violinplot() (with kind="violin")

boxenplot() (with kind="boxen")

**Categorical estimate plots:**

pointplot() (with kind="point")

barplot() (with kind="bar")

## Facet-Grid Plot

This class maps a dataset onto multiple axes arrayed in a grid of rows and columns that correspond to levels of variables in the dataset. The plots it produces are often called ‚Äúlattice‚Äù, ‚Äútrellis‚Äù, or ‚Äúsmall-multiple‚Äù graphics.

It can also represent levels of a third variable with the hue parameter, which plots different subsets of data in different colors. This uses color to resolve elements on a third dimension, but only draws subsets on top of each other and will not tailor the hue parameter for the specific visualization the way that axes-level functions that accept hue will.

**SYNTAX :-**

seaborn.FacetGrid(data, *, row=None, col=None, hue=None, col_wrap=None, sharex=True, sharey=True, height=3, aspect=1, palette=None, row_order=None, col_order=None, hue_order=None, hue_kws=None, dropna=False, legend_out=True, despine=True, margin_titles=False, xlim=None, ylim=None, subplot_kws=None, gridspec_kws=None)

```python
fg = sns.FacetGrid(tip,col="day",hue="sex",height=5)
fg.map(plt.scatter,"total_bill","tip").add_legend()
```

```python
fg = sns.FacetGrid(tip,col="day",hue="sex",height=5)
fg.map(plt.bar,"total_bill","tip").add_legend()
```
